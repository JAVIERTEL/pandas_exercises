### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://archive.ics.uci.edu/ml/machine-learning-databases/wine/wine.data). 

### Step 3. Assign it to a variable called wine

```python
data = {
    'Class': [1, 1, 1, 1, 1, 1],
    'Alcohol': [14.23, 13.20, 13.16, 14.37, 13.24, 14.20],
    'Malic_Acid': [1.71, 1.78, 2.36, 1.95, 2.59, 1.76],
    'Ash': [2.43, 2.14, 2.67, 2.50, 2.87, 2.45],
    'Alcalinity_of_Ash': [15.6, 11.2, 18.6, 16.8, 21.0, 15.2],
    'Magnesium': [127, 100, 101, 113, 118, 112],
    'Total_Phenols': [2.80, 2.65, 2.80, 3.85, 2.80, 3.27],
    'Flavanoids': [3.06, 2.76, 3.24, 3.49, 2.69, 3.39],
    'Nonflavanoid_Phenols': [0.28, 0.26, 0.30, 0.24, 0.39, 0.34],
    'Proanthocyanins': [2.29, 1.28, 2.81, 2.18, 1.82, 1.97],
    'Color_Intensity': [5.64, 4.38, 5.68, 7.80, 4.32, 6.75],
    'Hue': [1.04, 1.05, 1.03, 0.86, 1.04, 1.05],
    'OD280/OD315_of_Diluted_Wines': [3.92, 3.40, 3.17, 3.45, 2.93, 2.85],
    'Proline': [1065, 1050, 1185, 1480, 735, 1450]
}

wine = pd.DataFrame(data)
wine
```

### Step 4. Delete the first, fourth, seventh, nineth, eleventh, thirteenth and fourteenth columns

```python
# Eliminar las columnas especificadas
columns_to_drop = [0, 3, 6, 8, 10, 12, 13]
wine.drop(wine.columns[columns_to_drop], axis=1, inplace=True)
wine
```

### Step 5. Assign the columns as below:

The attributes are (donated by Riccardo Leardi, riclea '@' anchem.unige.it):  
1) alcohol  
2) malic_acid  
3) alcalinity_of_ash  
4) magnesium  
5) flavanoids  
6) proanthocyanins  
7) hue 

```python
new_column_names = ['alcohol', 'malic_acid', 'alcalinity_of_ash', 'magnesium', 'flavanoids', 'proanthocyanins', 'hue']
wine.columns = new_column_names
wine
```

### Step 6. Set the values of the first 3 rows from alcohol as NaN

```python
wine.loc[:2, 'alcohol'] = np.nan
wine
```

### Step 7. Now set the value of the rows 3 and 4 of magnesium as NaN

```python
wine.loc[3:4, 'magnesium'] = np.nan
wine
```

### Step 8. Fill the value of NaN with the number 10 in alcohol and 100 in magnesium

```python
# Llenar los valores NaN con 10 en la columna 'alcohol'
wine['alcohol'].fillna(10, inplace=True)

# Llenar los valores NaN con 100 en la columna 'magnesium'
wine['magnesium'].fillna(100, inplace=True)

# Mostrar el DataFrame resultante
wine
```

### Step 9. Count the number of missing values

```python
missing_values = wine.isnull().sum()
missing_values
```

### Step 10.  Create an array of 10 random numbers up until 10

```python
random_numbers = np.random.randint(0, len(wine), size=10)

```

### Step 11.  Use random numbers you generated as an index and assign NaN value to each of cell.

```python
wine.iloc[random_numbers] = np.nan
wine
```

### Step 12.  How many missing values do we have?

```python
total_missing_values = wine.isnull().sum().sum()
total_missing_values
```

### Step 13. Delete the rows that contain missing values

```python
wine.dropna(inplace=True)
wine
```

### Step 14. Print only the non-null values in alcohol

```python
# Step 14. Imprimir solo los valores no nulos en la columna 'alcohol'
print(wine['alcohol'].dropna())
```

### Step 15.  Reset the index, so it starts with 0 again

```python
wine.reset_index(drop=True, inplace=True)
wine
```
