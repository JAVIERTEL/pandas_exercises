### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### Step 2. Create the DataFrame that should look like the one below.

```python
# Crear el DataFrame con los datos proporcionados
data = {
    'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'],
    'last_name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze'],
    'age': [42, 52, 36, 24, 73],
    'female': [0, 1, 1, 0, 1],
    'preTestScore': [4, 24, 31, 2, 3],
    'postTestScore': [25, 94, 57, 62, 70]
}

df = pd.DataFrame(data)
df
```

### Step 3. Create a Scatterplot of preTestScore and postTestScore, with the size of each point determined by age
#### Hint: Don't forget to place the labels

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df['preTestScore'], df['postTestScore'], s=df['age']*10, alpha=0.5)

# Añadir etiquetas y título
plt.title('Scatterplot of preTestScore and postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')

# Mostrar el gráfico
plt.grid(True)
plt.show()
```

### Step 4. Create a Scatterplot of preTestScore and postTestScore.
### This time the size should be 4.5 times the postTestScore and the color determined by sex

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6)) #Tamaño del gráfico
scatter = plt.scatter(df['preTestScore'], df['postTestScore'],  #Datos a graficar
                      s=df['postTestScore']*4.5,  #Tamaño de los puntos
                      c=df['female'],  #Color de los puntos
                      cmap='bwr',  #Mapa de colores
                      alpha=0.5) #Transparencia

# Añadir etiquetas y título
plt.title('Scatterplot of preTestScore and postTestScore')
plt.xlabel('preTestScore')
plt.ylabel('postTestScore')

# Añadir una barra de color para indicar el sexo
cbar = plt.colorbar(scatter) #Añadir la barra de color
cbar.set_label('Female (1) / Male (0)') #Etiqueta de la barra de color

# Mostrar el gráfico
plt.grid(True)
plt.show()
```
