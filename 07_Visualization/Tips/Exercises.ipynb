### Step 1. Import the necessary libraries:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv). 

### Step 3. Assign it to a variable called tips

```python
url ="https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv"
tips = pd.read_csv(url)

tips.head()
```

### Step 4. Delete the Unnamed 0 column

```python
del tips['Unnamed: 0']

tips.head()
```

### Step 5. Plot the total_bill column histogram

```python
plt.figure (figsize=(10,6)) # tama√±o de la figura
plt.hist(tips['total_bill'], bins=30, color='blue', edgecolor='black') # histograma
plt.title('Total Bill Histogram') # titulo
plt.xlabel('Total Bill') # eje x
plt.ylabel('Frequency') # eje y
plt.grid(True) # cuadricula
plt.show() # mostrar
```

### Step 6. Create a scatter plot presenting the relationship between total_bill and tip

```python
#Crear el gr√°fico de dispersi√≥n

plt.figure(figsize=(10, 6))
plt.scatter(tips['total_bill'], tips['tip'], alpha=0.5)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.grid(True)
plt.show()
```

### Step 7.  Create one image with the relationship of total_bill, tip and size.
#### Hint: It is just one function.

```python
import seaborn as sns
plt.figure(figsize=(10, 6))
scatter = plt.scatter(tips['total_bill'], tips['tip'], s=tips['size']*50, alpha=0.5, c=tips['size'], cmap='viridis')
plt.title('Scatter Plot of Total Bill vs Tip with Size')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
sns.pairplot(tips)
```

### Step 8. Present the relationship between days and total_bill value

```python
# Crear el gr√°fico de caja utilizando seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='day', y='total_bill')
plt.title('Box Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.grid(True)
plt.show()

# Crear el gr√°fico de viol√≠n utilizando seaborn
plt.figure(figsize=(10, 6))
sns.violinplot(data=tips, x='day', y='total_bill')
plt.title('Violin Plot of Total Bill by Day')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.grid(True)
plt.show()
```

### Step 9. Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
# Crear el gr√°fico de dispersi√≥n utilizando seaborn
plt.figure(figsize=(10, 6))
sns.scatterplot(data=tips, x='tip', y='day', hue='sex', style='sex', s=100, palette='deep')
plt.title('Scatter Plot of Tip vs Day Differentiated by Sex')
plt.xlabel('Tip')
plt.ylabel('Day')
plt.grid(True)
plt.show()
```

### Step 10.  Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=tips, x='day', y='total_bill', hue='time', palette='Set3')
plt.title('Box Plot of Total Bill by Day Differentiated by Time')
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.legend(title='Time')
plt.grid(True)
plt.show()
```

### Step 11. Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
# Filtrar los datos para Dinner y Lunch
dinner_data = tips[tips['time'] == 'Dinner']
lunch_data = tips[tips['time'] == 'Lunch']

# Crear los histogramas uno al lado del otro
plt.figure(figsize=(14, 6))

# Histograma para Dinner
plt.subplot(1, 2, 1)
sns.histplot(dinner_data['tip'], bins=10, kde=False, color='blue')
plt.title('Histogram of Tip for Dinner')
plt.xlabel('Tip')
plt.ylabel('Frequency')

# Histograma para Lunch
plt.subplot(1, 2, 2)
sns.histplot(lunch_data['tip'], bins=10, kde=False, color='green')
plt.title('Histogram of Tip for Lunch')
plt.xlabel('Tip')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Step 12. Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
### They must be side by side.

```python
# Filtrar los datos para hombres y mujeres
male_data = tips[tips['sex'] == 'Male']
female_data = tips[tips['sex'] == 'Female']

# Crear los gr√°ficos de dispersi√≥n uno al lado del otro
plt.figure(figsize=(14, 6))

# Gr√°fico de dispersi√≥n para hombres
plt.subplot(1, 2, 1)
sns.scatterplot(data=male_data, x='total_bill', y='tip', hue='smoker', style='smoker', s=100, palette='coolwarm')
plt.title('Scatter Plot of Total Bill vs Tip (Male)')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

# Gr√°fico de dispersi√≥n para mujeres
plt.subplot(1, 2, 2)
sns.scatterplot(data=female_data, x='total_bill', y='tip', hue='smoker', style='smoker', s=100, palette='coolwarm')
plt.title('Scatter Plot of Total Bill vs Tip (Female)')
plt.xlabel('Total Bill')
plt.ylabel('Tip')

plt.tight_layout()
plt.show()
```
