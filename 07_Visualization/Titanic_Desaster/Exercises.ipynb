### Step 1. Import the necessary libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv)

### Step 3. Assign it to a variable titanic 

```python
url= "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Titanic_Desaster/train.csv"

titanic = pd.read_csv(url)

titanic.head()
```

### Step 4. Set PassengerId as the index 

```python
# Establecer 'PassengerId' como el Ã­ndice
titanic.set_index('PassengerId', inplace=True)

# Mostrar el DataFrame resultante
titanic.head()
```

### Step 5. Create a pie chart presenting the male/female proportion

```python
# Contar la cantidad de hombres y mujeres
sex_counts = titanic["Sex"].value_counts()
print(sex_counts)

# Crear el grÃ¡fico de pastel
plt.figure(figsize=(10, 5))  # TamaÃ±o del grÃ¡fico

# Crear el grÃ¡fico de pastel con mejoras
plt.pie(
    sex_counts,
    labels=sex_counts.index,
    autopct="%1.1f%%",
    startangle=140,
    colors=['#1f77b4', '#ff7f0e'],
    shadow=True,  # Agregar sombra
    explode=(0.05, 0.05),  # Separar las porciones
    wedgeprops={'edgecolor': 'black'}  # Agregar bordes
)

plt.title('DistribuciÃ³n de gÃ©nero en el Titanic', fontsize=16, fontweight='bold')  # TÃ­tulo del grÃ¡fico
plt.axis('equal')  # Hacer que el grÃ¡fico sea un cÃ­rculo
plt.show()  # Mostrar el grÃ¡fico
```

### Step 6. Create a scatterplot with the Fare payed and the Age, differ the plot color by gender

```python
# Crear el grÃ¡fico de dispersiÃ³n utilizando seaborn
plt.figure(figsize=(10, 6)) # TamaÃ±o del grÃ¡fico
sns.scatterplot(data=titanic, x='Fare', y='Age', hue='Sex', palette='coolwarm', s=100, alpha=0.7) # Crear el grÃ¡fico de dispersiÃ³n
plt.title('Scatter Plot of Fare vs Age Differentiated by Gender')
plt.xlabel('Fare')
plt.ylabel('Age')
plt.grid(True)
plt.show()
```

### Step 7. How many people survived?

```python
num_survived = titanic['Survived'].sum()
num_survived
```

### Step 8. Create a histogram with the Fare payed

```python
# Crear el histograma de la columna 'Fare'
plt.figure(figsize=(10, 6))
plt.hist(titanic['Fare'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Histogram of Fare')
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
