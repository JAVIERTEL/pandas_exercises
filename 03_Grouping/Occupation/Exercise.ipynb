### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
url= "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user"

users = pd.read_csv(url, sep='|', index_col='user_id')
users.head()
```

### Step 4. Discover what is the mean age per occupation

```python
users.groupby('occupation')["age"].mean()
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

```python
#Crear uan columna booleana para identificar a los hombres
users["is_male"] =users["gender"] == "M"

#Agrupar por ocupción y calcular la proporción de hombres

male_ratio_per_occupation = users.groupby ("occupation")["is_male"].mean()
male_ratio_per_occupation
#Ordenar de mayor a menor
sorted_male_ratio = male_ratio_per_occupation.sort_values(ascending=False)
sorted_male_ratio
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
#Agrupar por ocupación y calcular las edades max y min
min_max_age_per_occupation = users.groupby("occupation")["age"].agg(["min", "max"])
min_max_age_per_occupation
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
mean_age_per_occupation_gender = users.groupby(["occupation", "gender"])["age"].mean()
mean_age_per_occupation_gender
```

### Step 8.  For each occupation present the percentage of women and men

```python
#Contar el múmero de hombres y mujeres por ocupación
gender_occupations = users.groupby(["occupation", "gender"]).size().unstack()
gender_occupations

# Calcular el porcentaje de mujeres y hombres para cada ocupación
gender_percentage= gender_occupations.div(gender_occupations.sum(axis=1), axis=0) * 100

gender_percentage
```
